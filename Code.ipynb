```python
import tensorflow as tf
from tensorflow.keras import layers, models
import gradio as gr
import numpy as np
import PIL.ImageOps

# 1. LOAD DATA
print("Loading MNIST dataset...")
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 2. PREPROCESS (Reshape for CNN)
# CNNs expect a 4D shape: (Batch, Height, Width, Channels)
# Since it's grayscale, channels = 1
x_train = x_train.reshape(-1, 28, 28, 1) / 255.0
x_test = x_test.reshape(-1, 28, 28, 1) / 255.0

# 3. BUILD THE CNN ARCHITECTURE
model = models.Sequential([
    # First Convolutional Block: Finds basic edges
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)), # Shrinks data, keeps important parts

    # Second Convolutional Block: Finds complex shapes
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # The "Decision" Block
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5), # Stronger dropout for better generalization
    layers.Dense(10, activation='softmax')
])

# 4. COMPILE
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 5. TRAIN (10 Epochs for higher accuracy)
print("Training CNN...")
model.fit(x_train, y_train, epochs=10, validation_split=0.1)

# 6. UPDATED PREDICTION FUNCTION
def predict_digit(data):
    img = data['composite'].convert("L")
    img = PIL.ImageOps.invert(img)
    img = img.resize((28, 28))

    img_array = np.array(img) / 255.0
    # Reshape to match the CNN input (1, 28, 28, 1)
    img_array = img_array.reshape(1, 28, 28, 1)

    prediction = model.predict(img_array).tolist()[0]
    return {str(i): prediction[i] for i in range(10)}

# 7. LAUNCH
interface = gr.Interface(
    fn=predict_digit,
    inputs=gr.Sketchpad(type="pil"),
    outputs=gr.Label(num_top_classes=3),
    live=True,
    title="Pro Digit Classifier (CNN Edition)",
    description="This model uses Convolutional layers to achieve ~99% accuracy."
)

interface.launch(share=True)
```
